```python
def translate_categories(category_list):
    translated_categories = []
    for cat in category_list:
        cleaned_cat = cat.strip()
        if cleaned_cat in translation_dict:
            translated_categories.append(translation_dict[cleaned_cat])
        else:
            translated_categories.append(cleaned_cat)
            print(f"Категория '{cleaned_cat}' не найдена в словаре.")
    return translated_categories

df['translated_categories'] = df['categories'].apply(translate_categories)
print(df[['categories', 'translated_categories']].head(10))
```

```python
all_unique_categories = set()
for categories_list in df['translated_categories']:
    all_unique_categories.update(categories_list)

print(f"Количество уникальных меток: {len(all_unique_categories)}")
print("Уникальные метки:")
print(sorted(all_unique_categories))
```

```python
split_categories = df['translated_categories']
all_categories = [category.strip() for categories_list in split_categories for category in categories_list]
```

```python
from collections import Counter

category_counts = Counter(all_categories)
category_counts_df = pd.DataFrame.from_dict(category_counts, orient='index', columns=['count'])
category_counts_df = category_counts_df.sort_values(by='count', ascending=False)

print(category_counts_df.head(30))
```

```python
min_examples = 5000


popular_categories = {cat for cat, count in category_counts.items() if count >= min_examples}

print(f"Количество популярных меток: {len(popular_categories)}")
print("Популярные метки:", popular_categories)
```

```python
df['filtered_categories'] = df['translated_categories'].apply(
    lambda x: [cat for cat in x if cat in popular_categories]
)
```

```python
df_filtered = df[df['filtered_categories'].apply(len) > 0].reset_index(drop=True)

print(f"Исходный размер датасета: {len(df)}")
print(f"Размер датасета после фильтрации: {len(df_filtered)}")
```

```python
top_15_labels = ['Машинное обучение', 'Искусственный интеллект', 'Компьютерное зрение и распознавание образов',
                 'Вычисления и язык', 'Компьютеры и общество', 'Криптография и безопасность',
                 'Структуры данных и алгоритмы', 'Распределённые системы',
                 'Распределённые, параллельные и кластерные вычисления', 'Робототехника', 'Теория информации',
                 'Логика в информатике', 'Человеко-компьютерное взаимодействие', 'Информационный поиск']
```

```python
import ast

def clean_labels(row):
    if isinstance(row, list):
        labels = row
    else:
        try:
            labels = ast.literal_eval(row)
        except (ValueError, SyntaxError):
            labels = []

    filtered_labels = [label for label in labels if label in top_15_labels]
    return filtered_labels

df['filtered_labels'] = df['filtered_categories'].apply(clean_labels)
df = df[df['filtered_labels'].str.len() > 0]

df.head()
```

```python
label_mapping = {
    'Машинное обучение': 'Искусственный интеллект и машинное обучение',
    'Искусственный интеллект': 'Искусственный интеллект и машинное обучение',
    'Компьютерное зрение и распознавание образов': 'Компьютерное зрение и распознавание образов',
    'Вычисления и язык': 'Обработка естественного языка',
    'Структуры данных и алгоритмы': 'Фундаментальные основы CS',
    'Теория информации': 'Фундаментальные основы CS',
    'Логика в информатике': 'Фундаментальные основы CS',
    'Криптография и безопасность': 'Криптография и безопасность',
    'Распределённые, параллельные и кластерные вычисления': 'Распределённые и параллельные вычисления',
    'Человеко-компьютерное взаимодействие': 'Человеко-компьютерное взаимодействие',
    'Компьютеры и общество': 'Компьютеры и общество',
    'Информационный поиск': 'Информационный поиск',
    'Робототехника': 'Робототехника'
}

def map_labels(labels):
    mapped = []
    for label in labels:
        if label in label_mapping:
            mapped.append(label_mapping[label])
    return list(set(mapped))

df['grouped_labels'] = df['filtered_labels'].apply(map_labels)
df['main_label'] = df['grouped_labels'].apply(lambda x: x[0])
```
